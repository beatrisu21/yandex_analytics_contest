A. Довольные аналитики
В Яндексе есть два аналитика — Ян и Петр. Каждый день они берутся за новую задачу. Ян получает удовольствие от работы, если сделал 2 задачи подряд успешно. А Петр — только когда после проваленной задачи в следующей добился успеха.

На сколько дней в среднем Яну ждать первого удовольствия дольше, чем Петру (успех и провал по каждой задаче равновероятны)?




```python
!pip install matplotlib
import random
import matplotlib.pyplot as plt
```

```python
# Define a function to simulate Jan's problem-solving process
def jan_simulate_problem_solving():
    consecutive_solved = 0
    days = 0

    while consecutive_solved < 2:
        if random.random() < 0.5:
            consecutive_solved += 1
        else:
            consecutive_solved = 0
        days += 1

    return days

num_simulations = 100000
total_waiting_time = 0

# Lists to store waiting times from each simulation
waiting_times = []

# Run the simulation num_simulations times and store waiting times
for _ in range(num_simulations):
    waiting_time = yan_simulate_problem_solving()
    waiting_times.append(waiting_time)
    total_waiting_time += waiting_time

jan_average_waiting_time = total_waiting_time / num_simulations

# Plot a histogram of waiting times
plt.hist(waiting_times, bins=30, edgecolor='k')
plt.xlabel('Waiting Time (Days)')
plt.ylabel('Frequency')
plt.title('Distribution of Waiting Times to Solve 2 Problems in a Row')
plt.axvline(jan_average_waiting_time, color='r', linestyle='dashed', linewidth=2, label='Average Waiting Time')
plt.legend()
plt.show()

print(f"Average waiting time to solve 2 problems in a row: {jan_average_waiting_time} days")
```

```python
# Define a function to simulate 's problem-solving process
def peter_simulate_problem_solving():
    success = 0
    days = 0

    while success == 0:
        if random.random() < 0.5 and random.random() > 0.5 :
            success += 1
        days += 1

    return days

num_simulations = 100000
total_waiting_time = 0

# Lists to store waiting times from each simulation
waiting_times = []

# Run the simulation num_simulations times and store waiting times
for _ in range(num_simulations):
    waiting_time = peter_simulate_problem_solving()
    waiting_times.append(waiting_time)
    total_waiting_time += waiting_time

peter_average_waiting_time = total_waiting_time / num_simulations

# Plot a histogram of waiting times
plt.hist(waiting_times, bins=30, edgecolor='k')
plt.xlabel('Waiting Time (Days)')
plt.ylabel('Frequency')
plt.title('Distribution of Waiting Times to Solve 1 Problem after 1 Failed Problem')
plt.axvline(peter_average_waiting_time, color='r', linestyle='dashed', linewidth=2, label='Average Waiting Time')
plt.legend()
plt.show()

print(f"Average waiting time to solve 1 problem after 1 failed problem: {peter_average_waiting_time} days")
```

```python
# Answer
round(jan_average_waiting_time - peter_average_waiting_time) 
```
